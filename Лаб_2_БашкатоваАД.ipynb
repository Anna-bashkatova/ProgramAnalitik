# 2.1.14. 
Напишите программу на Python для отображения сетки и нарисуйте линейные графики значения закрытия AlphabetInc. в период с 3 октября 2016 года по 7 октября 2016 года. 

Настроил линии сетки с рендерингом с большей сеткой (основной сеткой) и меньшей сетка (второстепенная сетка). Включите сетку, но отключите галочки.

Фрагмент кода дает вывод, показанный на следующем снимке экрана:

image.png

```
import matplotlib.pyplot as plt
import datetime as dt
from matplotlib.dates import date2num

data = [(dt.datetime.strptime('2016-10-03', "%Y-%m-%d")),
        (dt.datetime.strptime('2016-10-04', "%Y-%m-%d")),
        (dt.datetime.strptime('2016-10-05', "%Y-%m-%d")),
        (dt.datetime.strptime('2016-10-06', "%Y-%m-%d")),
        (dt.datetime.strptime('2016-10-07', "%Y-%m-%d"))]

x = data
y = [772.559998,776.429993,776.469971,776.859985,775.080017]

fig = plt.figure() #область Figure (экземпляр класса figure)
ax = fig.add_subplot(111) #добавили к Figure область Axes

ax.plot(x,y,'r-o')
ax.set_xticks(x)
ax.set_xticklabels([date.strftime("%Y-%m-%d") for (date) in data])

plt.xlabel('Date') #Названия графика
plt.ylabel('Closing Value')
plt.title('Closing stock value of Alphabet Inc.') 

plt.minorticks_on() #Линии разметки
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle='-.', linewidth='0.5', color='black')

plt.show()
```

# 2.2.13. 
Напишите программу на Python для создания графиков с панелями ошибок на одном рисунке. Прикрепите текстовую метку над каждой строкой, отображая «мужчины» (целочисленное значение). 
Пример данных
Средняя скорость: 0,2474, 0,1235, 0,1737, 0,1824
Стандартное отклонение скорости: 0,3314, 0,2278, 0,2836, 0,2645.
Фрагмент кода дает вывод, показанный на следующем снимке экрана:

 image.png


```
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

N = 5
m = np.arange(N)
w = 0.35

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x=[54,42,67,58,30]
y=[4, 3, 4, 1, 5]
stolb1 = ax.bar(m, x, w, color='r', yerr=y)

def autolabel(rects):
  for rect in rects:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2., 1.05*height, '%d' % int(height),ha='center', va='bottom')
autolabel(stolb1)

red_patch = mpatches.Patch(color='r', label='Men')
plt.legend(handles=[red_patch])

plt.xlabel('Velocity') #Названия графика
plt.ylabel('Scores')
plt.title('Scores by Velocity') 

plt.show()
```

# 2.4.5. 
Напишите программу на Python, чтобы нарисовать точечный график для трех разных групп с разной массой и высотой. 
Фрагмент кода дает вывод, показанный на следующем снимке экрана:
image.png

```
import matplotlib.pyplot as plt
import numpy as np 

fig = plt.figure()
ax = fig.add_subplot(111)

weight1=[67,57.2,59.6,59.64,55.8,61.2,60.45,61,56.23,56]
height1=[101.7,197.6,98.3,125.1,113.7,157.7,136,148.9,125.3,114.9] 

weight2=[61.9,64,62.1,64.2,62.3,65.4,62.4,61.4,62.5,63.6]
height2=[152.8,155.3,135.1,125.2,151.3,135,182.2,195.9,165.1,125.1] 

weight3=[68.2,67.2,68.4,68.7,71,71.3,70.8,70,71.1,71.7]
height3=[165.8,170.9,192.8,135.4,161.4,136.1,167.1,235.1,181.1,177.3]

weight=np.concatenate((weight1,weight2,weight3))
height=np.concatenate((height1,height2,height3))

plt.scatter(weight, height, marker = '*')
plt.xlabel('weight')
plt.ylabel('height')
plt.title('Group wise Weight vs Height scatter plot')

plt.show()
```
